```python
import numpy as np

#Creating a random 1 rank array of size 15
Random=np.random.randint(20, size=15)

#Printing the array
print("Output Array:", Random)

#Re-shaping array into 3*5 matrix using reshape()
Array_Shape=Random.reshape(3,5)
print("Re-shaped Array:\n",Array_Shape)

#Printing the highest number in each row
c=np.amax(Array_Shape,axis=1)
print("Maximum value in each row:", c)


#Using np.where(),printing the maximum value in each row
Max_value=np.where(Array_Shape==np.amax(Array_Shape,axis=1,keepdims=True),0,Array_Shape)
print(Max_value)

#A Diagonal is extracted from the array
Diagonal=np.diag(Array_Shape)
print(Diagonal)

#Desired output is peinted in the output file.
Resultant=np.savetxt("output.npy",Diagonal)
print(Resultant)
```

```python
# imported pandas library 
import pandas as pd
import os
# imported matplotlib library
import matplotlib.pyplot as plot
# Input csv file is read by using read_csv() 
df = pd.read_csv("C:\\Users\\laliswathi\\downloads\\data.csv")
# describe() is used to  give brief describtion of the data frame
print('Brief description of data frame')
print(df.describe())
# To check if the dataframe contains null values and replace the null values with  mean of that column
print("The count of null values in the dataframe",df.isnull().values.any().sum())
df = df.fillna(value=df.mean())
print("The count of null values in the dataframe",df.isnull().values.any().sum())
#To select atleast 2 columns aand aggregate the data     
print('Using Aggregate functions on columns')
print(df[['Duration','Calories','Pulse','Maxpulse']].agg(['min','mean','max','count']))
#To filter the data frame to select rows with calories between 500,1000
print(df[ df['Calories'].between(500,1000)])
#To filter the data frame to select rows with calories > 500 and pulse<100
print(df[ (df['Calories'] >500) &  (df['Pulse']<100)])
#Created new modified dataframe that contains all columns except Maxpulse
df_modified = df.loc[ : , df.columns != 'Maxpulse']
print(df_modified)
#Deleted maxpulse column from main df dataframe
del df['Maxpulse']
print(df)
#Converted datatype of calories column to int datatype
df["Calories"] = df["Calories"].round(0).astype('int')
print(df.dtypes)
print(df)
dataFrame = pd.DataFrame(data=df, columns=['Calories','Duration']);

# TO Draw a scatter plot between Duration and Calories columns 

dataFrame.plot.scatter(x='Duration', y='Calories', title= "Scatter plot between two variables Calories and Duration");
plot.show(block=True);
```
